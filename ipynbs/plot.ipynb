```python
import os; 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dirnames = "baseline_model/  baseline_model_pretrained/  revised_model/  revised_model_pretrained/  wordLevel_model/  wordLevel_model2/  wordLevel_model2_lateral/".split()
```

```python
plt.figure(figsize=(10,10))

for dirn in dirnames:
    filename = os.path.join(os.getcwd(), 'saved_exp/', dirn, 'history.csv')
    history = pd.read_csv(filename)
    epochs = list(range(10))
#     plt.plot(epochs, history.train_o_loss, label=f'{dirn} train cross entropy loss')
#     plt.plot(epochs, history.valid_o_loss, label=f'{dirn} valid cross entropy loss')
    if 'revised' in dirn:
        pass
#         plt.plot(epochs, history.train_c_loss/10, label=f'{dirn} train classification loss')
        plt.plot(epochs, history.valid_c_loss/10, label=f'{dirn} val classification loss')
    else:
        pass
#         plt.plot(epochs, history.train_c_loss, label=f'{dirn} train classification loss')
        plt.plot(epochs, history.valid_c_loss, label=f'{dirn} val classification loss')
    plt.legend()

n = 'valid classification loss'
plt.title(n, loc='center')
plt.savefig(f'save/{n}')
```
